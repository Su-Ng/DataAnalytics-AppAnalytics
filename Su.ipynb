 # Read the file in CSV format

```python
import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv('googleplaystore.csv')
```

```python
df.info()
```

# Select column Category and Rating

```python

df[['Category','Rating']]
```

# Remove rows with NA

```python
df=df.dropna()
```

```python
df[['Category','Rating']]
```

# Find the total number of apps in each category

```python
df['Category'].value_counts()






```

# Plot the number of apps in each category

```python
ax=df['Category'].value_counts().plot(kind='bar', grid=True, figsize=(10,5))
ax.legend(["category count"])
```

```python
range = (1, 5)
bin_list=[1.5,2.5,3.5,4.5,5.5]
```

# plot out the different app rating (range)

```python
plt.hist(df['Rating'].values, bins=bin_list, edgecolor="k",rwidth = 0.8)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
#plt.figure(figsize=[20,5])
plt.rcParams["figure.figsize"] = [10,4]
plt.legend(["category count"])
```

```python
df.groupby(['Category']).mean().plot(kind='bar', grid=True,  figsize=(12,6))
```

# Find the avg rating of each category

```python
df.groupby(['Category']).mean()
    
```

# Find the category with the highest rating

```python
df2=df.groupby(['Category']).mean().sort_values('Rating')
```

```python
print(df2.tail(n=1))
```
